```python
#Datei lesen
lines = []
import csv
with open('Rezensionen_Schlüsselroman.csv', newline='', encoding="UTF8") as csvfile:
    datumreader = csv.reader(csvfile, delimiter=';', quotechar='|')
    for row in datumreader:
        #print(', '.join(row))#print(row)
        #print(', '.join(row))
        lines.append(', '.join(row))
lines
```

```python
#Datum extrahieren

matches = {}
p = re.compile('(?P<day>\d{1,2}.)\s??(?P<month>\d{1,2}.|[a-z,A-Z,ä,ü,ö]{1,13})\s??(?P<year>\d{4})')
i=0

for line in lines:
    
    #print(line)
    m = p.search(line)
    if m:
        i += 1
        Tag, Monat, Jahr = m.group(1), m.group(2), m.group(3)
        
        #print("Tag", m.group(1), "Monat", m.group(2), "Jahr", m.group(3))
        #age = int(m.group(3)) - int(m.group(2))
        #matches[m.group(1)] = [m.group(2), m.group(3), age]
        matches[i] = line, Tag, Monat, Jahr
    else:
        pass
        #matches[i] = line
    
for key in matches.keys():
    line, Tag, Monat, Jahr = matches[key]
    print(key, line[:200])
    
```

```python
#erste Rez suchen
import datetime
from datetime import date
import unicodedata as ud


dateswalser = []
datesbiller = []
datesgstrein = []
Monate = ["Januar", "Februar", "März", "April", "Mai", "Juni", "Juli", "August", "September", "Oktober", "November", "Dezember"]
j = 0
for rez in matches.keys():
    j +=1
    
        
    try:
        line, Tag, Monat, Jahr = matches[rez]
        Tag = Tag.strip(".")
        Monat = Monat.strip(".")
        Monat = ud.normalize('NFC',Monat)
        i=0
        for mon in Monate:
            i += 1
            if mon.startswith(Monat):
                Monat = str(i)
                #print(Monat)
        
        Jahr = Jahr.strip(".")
        if j <= 61:
            dateswalser.append('-'.join([Tag, Monat, Jahr]))
        elif j<=82:
            datesbiller.append('-'.join([Tag, Monat, Jahr]))
        else:
            datesgstrein.append('-'.join([Tag, Monat, Jahr]))
    except:
        pass
    

walsermin= sorted(dateswalser, key=lambda x: datetime.datetime.strptime(x, '%d-%m-%Y'))[0]
print(walsermin)
billermin= sorted(datesbiller, key=lambda x: datetime.datetime.strptime(x, '%d-%m-%Y'))[0]
print(billermin)
gstreinmin= sorted(datesgstrein, key=lambda x: datetime.datetime.strptime(x, '%d-%m-%Y'))[0]
print(gstreinmin)

```

```python
# Tage seit Erstrezension berechenen
matches = {}
j = 0
for line in lines:
    j +=1
    i=0
    #print(line)
    m = p.search(line)
    if m:
        Tag, Monat, Jahr = m.group(1).strip("."), m.group(2).strip("."), m.group(3).strip(".")
        i=0
        for mon in Monate:
            i += 1
            if mon.startswith(Monat):
                Monat = str(i)
        #print(Monat)
        
                
        a = date(int(Jahr),int(Monat),int(Tag))
        print("a", a)
        if j<=61:
            min = walsermin.split("-")
            Tag2 = int(min[0])
            Monat2 = int(min[1])
            Jahr2 = int(min[2])
            b= date(Jahr2, Monat2, Tag2)
        elif j <= 82:
            min = billermin.split("-")
            Tag2 = int(min[0])
            Monat2 = int(min[1])
            Jahr2 = int(min[2])
            b= date(Jahr2, Monat2, Tag2)
        else:
            min = gstreinmin.split("-")
            Tag2 = int(min[0])
            Monat2 = int(min[1])
            Jahr2 = int(min[2])
            b= date(Jahr2, Monat2, Tag2)
        print("min", b)
        Tage = (a-b).days
        print("diff:", Tage)
        #print("Tag", m.group(1), "Monat", m.group(2), "Jahr", m.group(3))
        #age = int(m.group(3)) - int(m.group(2))
        #matches[m.group(1)] = [m.group(2), m.group(3), age]
        matches[line] = Tag, Monat, Jahr, Tage
    else:
        matches[line] = ""
```
